```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_theme(style="ticks")
```

```python
def rzuc_moneta():
    moneta = ['orzel', 'reszka']
    return moneta[np.random.randint(0,2)]

uczestnicy = pd.DataFrame(columns=['ekstrawersja', 'grupa'])
```

```python
# dodaj uczestnika
uczestnicy = uczestnicy.append({'ekstrawersja': float(np.random.normal(100, 15, 1)), 'grupa': rzuc_moneta()}, ignore_index=True)
uczestnicy
```

```python
# pokaż grupę
sns.catplot(x='grupa', y='ekstrawersja', data=uczestnicy, kind='swarm')

# oblicz średnie
print(uczestnicy.groupby('grupa').ekstrawersja.mean())
print(uczestnicy.groupby('grupa').ekstrawersja.std())
```

```python
# dodaj uczestnika x 10
for x in range(100):
    uczestnicy = uczestnicy.append({'ekstrawersja': float(
        np.random.normal(100, 15, 1)), 'grupa': rzuc_moneta()}, ignore_index=True)
```
